# Toronto Neighbourhoods

### Step 1: Scraping the data from Wikipedia
#### Method using 'Beautiful Soup'

```python
import urllib.request
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
page = urllib.request.urlopen(url)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
all_tables=soup.find_all("table")
all_tables
```

```python
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

```python
A=[]
B=[]
C=[]
for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
# Assigning Columns
import pandas as pd
df=pd.DataFrame(A,columns=['Postal_Code'])
df['Borough']=B
df['Neighbourhood']=C
```

```python
df.head(5)
```

### Cleaning the data 
#### Removing \n and Borough 'Not assigned'

```python
df['Postal_Code'] = df['Postal_Code'].replace('\n','', regex=True)
df['Borough'] = df['Borough'].replace('\n','', regex=True)
df['Neighbourhood'] = df['Neighbourhood'].replace('\n','', regex=True)

```

```python
df = df[df.Borough != 'Not assigned']
```
